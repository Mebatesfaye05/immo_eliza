```python
#import 
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectFromModel
```

```python
# Read the csv file
data_file = r'C:\Users\Gebruiker\Downloads\data_a.csv'
data = pd.read_csv(data_file)
print(data)
```

```python
# drop duplicated
data=(data.drop_duplicates())
print(data)

```

```python
## drop NAN
data = data.dropna()
print(data)
```

```python
# select only numeric columns
data = data.select_dtypes(exclude='object')
print(data)
```

```python
# drop specified columns
columns_to_drop = ['Open fire', 'Unnamed: 0','Surface of the land', 'Swimming pool']
data = data.drop(columns=[col for col in columns_to_drop if col in data.columns])
print(data)
```

```python
#calculating correlation
corr = data.corr()
print(corr)
```

```python
#Normalizing the data 
price = data["Price"]
scaler = StandardScaler()
data[data.columns] = scaler.fit_transform(data[data.columns])
data["Price"] = price
print(data)
```

```python
# Split data into features and target variable
X = data[['Number of rooms','Living Area', 'Terrace', 'Area of the terrace', 'Area of the garden', 'zipcode','Surface area of the plot of land','Number of facades']]
y = data['Price']
```

```python
# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2)
```

```python
# create a model
rf = RandomForestRegressor(n_estimators=300, max_depth = 250)
# Fit the model to the training data
rf.fit(X_train, y_train)


```

```python
# Make predictions on the test data
y_pred = rf.predict(X_test)
#mean of price
print("This is the mean price of a house", data["Price"].mean())
# calculate accurracy before using mean
print("This is the accurracy of the data before using mean Price of a house :", rf.score(X_test, y_test))
#mean absolute error
error =abs(y_pred - y_test).mean() 
print("This is the mean absolute error of the data : ",error)
#accurracy of the model 
print("This is the accuracy of the data after using the mean price of a house: ", abs(1-error/data["Price"].mean())*100,"%")
```
